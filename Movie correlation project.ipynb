```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\anilk\Downloads\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = round(np.mean(df[col].isnull()),2)
    print('{} - {}%'.format(col , pct_missing))
```

```python
df.isnull().sum()
```

```python
df = df.fillna({'budget': 0, 
                'gross': 0
               })
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df.sort_values( by = 'gross' , inplace = False , ascending = False)
```

```python
#scatter plot using budget and gross;
plt.scatter(x = 'budget' , y = 'gross' , data = df)
plt.title('Budget vs gross')
plt.xlabel('budget')
plt.ylabel('Gross')
plt.show()
```

```python
#regplot with gross and budget
sns.regplot(x = 'budget' , y = 'gross' , data = df, scatter_kws = {"color" : "red"} , line_kws = {"color":"blue"})
```

```python
corelation_matrix = df.corr(method = 'pearson' , numeric_only = True )
```

```python
sns.heatmap( corelation_matrix , annot = True)
plt.title('corelation matric for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df_numerized = df
```

```python
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
sns.heatmap(df_numerized , annot = True)
plt.show()
```

```python
corealtion_matrix = df_numerized.corr()
co_pairs = corealtion_matrix.unstack()
co_pairs
```

```python
sorted_pairs = co_pairs.sort_values(ascending = True)
```

```python
sorted_pairs_sorted =  sorted_pairs[(sorted_pairs) > 0.5]
```
