```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"Customer Churn.csv")
```

```python
df.head(30)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(" " , "0")
df['TotalCharges'] = df['TotalCharges'].astype(float)
df.info()
```

```python
df['customerID'].duplicated().sum()
```

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"
```

```python
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
df.head(30)
```

```python
ax = sns.countplot(x = 'SeniorCitizen' , data = df , hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb= df.groupby('Churn').agg({'Churn': 'count'})
#gb
plt.pie(gb['Churn'], labels = gb.index, autopct='%2.1f%%')
plt.title("% of churn customer" , fontsize = 10)
plt.show()
```

```python
# from the given pie chart we can conclude that 26 % of customers have churned out
```

```python
plt.figure(figsize = (3, 4))
sns.countplot(x = 'gender' , data = df, hue = 'Churn')
plt.title('Churned on the basis of gender')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the columns you want to plot
cols = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the figure and axes
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # 3x3 grid
axes = axes.flatten()  # Flatten to make it iterable

# Loop through each column and create a countplot
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], hue = 'Churn')
    axes[i].set_title(f'Count of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 3 , ncols = 3 , figsize= (15,15))
axes = axes.flatten()

for i , col in enumerate(cols):
  ax = sns.countplot(x= col , ax= axes[i] , data = df , hue = 'Churn')
  ax.set_title(f'count of {col}')
  ax.set_xlabel(col)
  ax.set_ylabel('count')

plt.tight_layout()
plt.show()




```
